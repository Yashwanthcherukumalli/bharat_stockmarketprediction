```python
# Time Series Analysis - Closing Prices Over Time
df['TIMESTAMP'] = pd.to_datetime(df['TIMESTAMP'])
plt.figure(figsize=(12, 6))
plt.plot(df['TIMESTAMP'], df['CLOSE'])
plt.title('Closing Prices Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Closing Price')
plt.show()
```

```python
# Volatility Analysis - Bollinger Bands
window = 20
df['RollingMean'] = df['CLOSE'].rolling(window=window).mean()
df['UpperBand'] = df['RollingMean'] + 2 * df['CLOSE'].rolling(window=window).std()
df['LowerBand'] = df['RollingMean'] - 2 * df['CLOSE'].rolling(window=window).std()

plt.figure(figsize=(12, 6))
plt.plot(df['TIMESTAMP'], df['CLOSE'], label='Closing Price', color='blue')
plt.plot(df['TIMESTAMP'], df['RollingMean'], label='Rolling Mean', color='orange')
plt.plot(df['TIMESTAMP'], df['UpperBand'], label='Upper Band', linestyle='dashed', color='red')
plt.plot(df['TIMESTAMP'], df['LowerBand'], label='Lower Band', linestyle='dashed', color='green')
plt.fill_between(df['TIMESTAMP'], df['UpperBand'], df['LowerBand'], alpha=0.2, color='gray')
plt.title('Bollinger Bands for Closing Prices')
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.legend(loc='best')
plt.show()
```

```python
# Moving Average Crossover Strategy
short_window = 10
long_window = 50

df['ShortMA'] = df['CLOSE'].rolling(window=short_window, min_periods=1).mean()
df['LongMA'] = df['CLOSE'].rolling(window=long_window, min_periods=1).mean()

plt.figure(figsize=(12, 6))
plt.plot(df['TIMESTAMP'], df['CLOSE'], label='Closing Price', color='blue')
plt.plot(df['TIMESTAMP'], df['ShortMA'], label=f'{short_window}-Day MA', color='orange')
plt.plot(df['TIMESTAMP'], df['LongMA'], label=f'{long_window}-Day MA', color='green')
plt.title('Moving Average Crossover Strategy')
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.legend(loc='best')
plt.show()

```

```python
scaler = MinMaxScaler(feature_range = (0,1))
df_open_scale = scaler.fit_transform(np.array(df_open).reshape(-1,1))
df_open_scale.shape
```

```python
training_size = int( len(df_open_scale) * 0.75)
testing_size  = len(df_open_scale) - training_size
train_data , test_data = df_open_scale[0 : training_size, :], df_open_scale[training_size : len(df_open_scale), :1]
training_size , testing_size
```

```python
def create_dataset(dataset, time_step=1):
    train_X, train_Y = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]
        train_X.append(a)
        train_Y.append(dataset[i + time_step, 0])
    return np.array(train_X), np.array(train_Y)
```

```python
time_step = 100
x_train, y_train = create_dataset(train_data, time_step)
x_test, y_test = create_dataset(test_data, time_step)
```

```python
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1], 1)
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1], 1)
```

```python
print(x_train.shape), print(y_train.shape)
```

```python
model = Sequential()
model.add(LSTM(500, return_sequences = True, input_shape = (1000,1)))
model.add(LSTM(500, return_sequences = True))
model.add(LSTM(500))
model.add(Dense(100))
model.compile(loss = 'mean_squared_error', optimizer = 'adam')
model.summary()
```

```python
model.fit(x_train, y_train, validation_data = (x_test, y_test), epochs = 10, batch_size = 5000, verbose = 1)
```

```python
#Lets predict and check performance metrics

train_predict = model.predict(x_train)
test_predict = model.predict(x_test)
```
